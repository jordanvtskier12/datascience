```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None   #default='warn'
```

```python
df = pd.read_excel("http://bit.ly/2VuOA98")
dfhours = df[['Watching_TV_Hours', 'Doing_Homework_Hours']]
dfhours.columns = ['tv','homework']
```

This dataset has many rows (one for each question), so we simplified it by stripping all rows except for those dealing with hours spent on TV and HW

```python
print(dfhours.tv.unique())
print(dfhours.homework.unique())

dfhours.tv = pd.to_numeric(dfhours.tv, errors='coerce')
dfhours.homework = pd.to_numeric(dfhours.homework, errors='coerce')
```

Check for and remove non-number values

```python
plt.scatter(dfhours.tv, dfhours.homework)
plt.show()
```

plot data to check for outliers

```python
dfhours = dfhours.dropna()

plt.hist(dfhours.tv)
plt.show()
```

```python
plt.hist(dfhours.tv)
plt.title('TV Hours')
plt.show()
plt.hist(dfhours.homework)
plt.title('Homework Hours')
plt.show()
```

```python
dfhours = dfhours[dfhours.tv <= 20]
dfhours = dfhours[dfhours.homework <= 30]
```

Filter data to remove outliers 

```python
plt.hist(dfhours.tv)
plt.title('TV Hours')
plt.show()
plt.hist(dfhours.homework)
plt.title('Homework Hours')
plt.show()
```

Plot data again to check for errors
