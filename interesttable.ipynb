# Interest Table
This is an example and use of the *interest equation*: $ A=P * (1 +\frac {r} {n})^{t}$ 

```python
df = pd.DataFrame()
df['years'] = range(31)
df['value_float'] = [1000 * (1.035) ** i for i in df.years]
df['value_str'] = ["$" + str(int(1000 * (1.035) ** i)) for i in range(31)]
df
```

```python
plt.plot(df.years, df.value_float)
plt.xlabel(df.years.name)
plt.ylabel(df.value_float.name)
plt.title("Money in account with 3.5% interest on 1000$ initial over 30 years")
plt.legend()
plt.show()
```
