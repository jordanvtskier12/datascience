# Ballistics and the Range Equation
This is an ivestigation of the so-called *range equation*: $d=\frac {v^2} g\sin 2\theta$

```python
from math import radians, sin, pi
from numpy import linspace
import matplotlib.pyplot as plt
```

```python
def rangeformula(v, theta):
    return (v**2/9.81)*sin(2*theta)
```

```python
angledegrees = range(0, 91)
```

```python
rangefromdegrees100 = [rangeformula(100, radians(angle)) for angle in angledegrees]
```

```python
plt.plot(angledegrees, rangefromdegrees100)
plt.title('Range as a function of angle at 100 m/s')
plt.xlabel('Angle (degrees)')
plt.ylabel('Range(meters)')
plt.show()
```

```python
rangefromdegrees200 = [rangeformula(200, radians(angle)) for angle in angledegrees]
```

```python
plt.plot(angledegrees, rangefromdegrees100, label = "100 m/s")
plt.legend()
plt.plot(angledegrees, rangefromdegrees200, label = "200 m/s")
plt.legend()
plt.title('Range as a function of angle at various velocities')
plt.xlabel('Angle (degrees)')
plt.ylabel('Range (meters)')
plt.show()
```

```python
velocities = [(v) for v in range(100)]
```

```python
rangevelocities = [rangeformula(v , 45) for v in velocities]
```

```python
plt.plot(velocities, rangevelocities, label= str(45)+"°")
plt.legend()
plt.title('Range from Initial Velocity at 45°')
plt.xlabel('Velocity (meters/second)')
plt.ylabel('Range (meters)')
plt.show()
```

```python
for k in range(1,6):
    rangevelocities = [rangeformula(v , radians(k * 15) ) for v in velocities]
    plt.plot(velocities, rangevelocities, label= str(k * 15)+"°")
    plt.legend()
plt.title('Range from Initial Velocity at Various Angles')
plt.xlabel('Velocity (meters/second)')
plt.ylabel('Range (meters)')
plt.show()
```
